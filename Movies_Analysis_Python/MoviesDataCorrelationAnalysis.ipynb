```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the configuration of the plots we will create


# Read in the data

df = pd.read_csv(r'C:\Users\chans\Downloads\archive\movies.csv')
```

```python
# Look at the data

df.head()
```

```python
# Looking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))

# Dropping missing data for now
df = df.dropna()
```

```python
# Data types for columns

df.dtypes
```

```python
# Changing data types of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df
```

```python
# Create correct Year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# To show all rows
pd.set_option('display.max_rows', None)
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False).head()
```

```python
# Check for duplicates

df['company'].drop_duplicates().sort_values(ascending=False).head()
```

```python
# Budget probably has high correlation
# Company probably has high correlation


# Scatter plot with budget vs gross revenue
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# translate company names with numerical value, so convert names to numbers

df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head() # all non-numeric fields have been translated into numeric
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation for Numeric Fields')
plt.xlabel('Field')
plt.ylabel('Field')

plt.show()
```

```python
# Looks like Company and gross are not as correlated
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack() # lists correlation of field with every other field

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5] # shows pairs of fields with higher corr

high_corr.head()
```
